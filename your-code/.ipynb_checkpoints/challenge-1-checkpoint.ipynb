### Cleanup `Name` that contain "Mega"

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
def cleanning_mega(string):
    if 'Mega' in string:
        return 'Mega' + ' '.join(string.split(' ')[1:])
    else:
        return string

pokemon['Name'] = pokemon['Name'].agg(cleanning_mega)

# test transformed data
pokemon.head(10)
```

### Create a new column called `A/D Ratio` whose value equals to `Attack` devided by `Defense`

For instance, if a pokemon has the Attack score 49 and Defense score 49, the corresponding `A/D Ratio` is 49/49=1.

```python
# enter your code here
pokemon['A/D Ratio'] = pokemon['Attack'] / pokemon['Defense']
pokemon
```

### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
#Se comprueba que en la columna Type 2 hay 386 NaN
pokemon.isnull().sum()
```

```python
# enter your code here

#Primero se limpian los NaN volviendo a crear las columnas Type 1 y Type 2, asignÃ¡ndoles .fillna
#Luego para crear la columna Combo Type se utiliza where de numpy para especificar la uniÃ³n de Type 1 y Type 2.

pokemon['Type 1'] = pokemon['Type 1'].fillna('')
pokemon['Type 2'] = pokemon['Type 2'].fillna('')
pokemon['Combo Type'] = np.where((pokemon['Type 1'] != ''), pokemon['Type 1'] + pokemon['Type 2'], pokemon['Type 2'])
pokemon
```

### Identify the pokemons whose `A/D Ratio` are among the top 5 

```python
# enter your code here

#Se utiliza sor_values para ordenar de mayor a menor. 
top_five = pokemon.sort_values('A/D Ratio', ascending= False).head()
top_five
```

### For the 5 pokemons printed above, aggregate `Combo Type` and use a list to store the unique values.

Your end product is a list containing the distinct `Combo Type` values of the 5 pokemons with the highest `A/D Ratio`.

```python
# enter your code here

#Se crea una nueva lista para almacenar el contenido de la columna Comobo Type.
#Se utiliza la lista anterior top_five que contiene los valores mÃ¡s altos de la columna A/D Ratio.
#Se especifica que de la columna 'Combo Type', de nuestra lista top_five haremos un set del contenido que se pasa a un arreglo de una lista.

list_combo = list(set(top_five['Combo Type']))
list_combo
```

### For each of the `Combo Type` values obtained from the previous question, calculate the mean scores of all numeric fields across all pokemons.

Your output should look like below:

![Aggregate](aggregated-mean.png)

```python
# enter your code here

#Se crea una lista final.
#Se agrupa la columna Combo Type para sacarle la media de los valos res de la columna
#Luego se indica que la final_list estarÃ¡ ordenada segÃºn con los datos de la previa lista list_combo. 


final_list = pokemon.groupby(['Combo Type']).mean().transpose()
final_list = final_list[list_combo].transpose()
final_list
```
